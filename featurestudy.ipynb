# Win rate por cada jogada feita ( hit / stand )

```python
# Group by 'action' and calculate count and mean of 'rewards'
grouped_data = df.groupby('action')['rewards'].agg(['count', 'mean']).reset_index()

# Rename the columns for clarity
grouped_data.rename(columns={'count': 'num_episodes', 'mean': 'win_mean'}, inplace=True)

# Calculate win_rate using the provided formula
grouped_data['win_rate'] = ((grouped_data['win_mean'] + 1) / 2) * 100

# Display the grouped data
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(grouped_data)
```

```python
action_count = df.groupby(["obs", "action"]).size().reset_index(name='count_action')

# Display the entire DataFrame without truncation
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(action_count)
```

```python
grouped_data = df.groupby(["obs", "rewards"]).size().reset_index(name='count_action')

# Display the entire DataFrame without truncation
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(grouped_data)
```

```python
# Assuming df1 and df2 are your two dataframes
# Assuming 'common_column' is the common column based on which you want to merge

merged_df = pd.merge(action_count, grouped_data, on='obs', how='outer')

merged2 = merged_df
merged2['rate'] = (merged_df['count_action_y']/merged_df['count_action_x'])*100
merged2 = merged2[(merged2['rate'] < 100)]  #havia alguns outliers que estragavam os resultados
```

```python
merged2= merged2.drop('count_action_x', axis=1)
merged2 =merged2.drop('count_action_y', axis=1)
```
